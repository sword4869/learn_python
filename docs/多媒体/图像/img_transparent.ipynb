```python
import imageio.v3 as iio
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = iio.imread('/home/sword/Downloads/output-onlinepngtools.png')
```

```python
img.shape       # [H, W, C]
```

```python
plt.imshow(img)
```

```python
img = np.array(img) / 255.0
# 
# [H, W, 3] * [H, W, 1] + [H, W, 1] = [H, W, 3]
# (1.0 - img[..., -1:])  (1.0 - img[..., -1:]) * 1, 1nparrayrgb255
img2 = img[..., :3] * img[..., -1:] + (1.0 - img[..., -1:])
plt.imshow(img2)
```
